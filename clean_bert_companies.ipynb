```python
import pandas as pd
from collections import Counter
import clean_df as cd
```

```python
bert_set_2017 = pd.read_csv('DistilBERT_Pharma_Companies_2017.csv')
```

```python
def majority_rules(*args):
    c = Counter(list(args))
    value, count = c.most_common()[0]
    return value
```

```python
majority_rules('bob', 'neil', 'alan')
```

```python
bert_set_2017.head()
```

```python
bert_set_2017['bert_final'] = bert_set_2017.apply(lambda x: majority_rules(x['Johnson & Johnson'], x['Johnson & Johnson.1'], x['Johnson & Johnson.2']), axis=1)
```

```python
df = cd.clean(bert_set_2017, 'bert_final', True)
```

```python
print(len(df))
df = df[['bert_final', 'name', 'clean', '-company', '-states', '-common', 'short']]
df['len'] = df['name'].map(len)
df['cl-len'] = df['clean'].map(len) # remove things that leave nothing after being 'cleaned'

pd.set_option('display.expand_frame_repr', False)
print(df.columns)
df = df[df['cl-len'] > 0] 
pd.set_option('display.max_rows', 100) # this display will not work, if the value is less than the number in the data frame
display(df[df['cl-len'] < 5].head(100))
```
